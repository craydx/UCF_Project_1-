```python
# sample data to analysis - state = AK year = 2020 to compare against data shown on website

#Build query URL with appropriate filters to compare against website data
base_url="https://chronicdata.cdc.gov/resource/f8ti-h92k.json"
yrstrt=2020
yrend=2020
locabbr="AK"
strat1="White, non-Hispanic"
questid="DIA1_1"
filters=f"locationabbr={locabbr}&yearstart={yrstrt}&yearend={yrend}&stratification1={strat1}&questionid={questid}"
query_url = f"{base_url}?{filters}&$$app_token={api_token}"

#Get data and convert to data frame
sample_data = requests.get(query_url).json()

df=pd.DataFrame(sample_data)
df
```

```python
# find all question id's available, stratificationcategories/id... pick state = CA

base_url="https://chronicdata.cdc.gov/resource/f8ti-h92k.json"
#yrstrt=2020
#yrend=2020
locabbr="CA"
limit=50000
filters=f"locationabbr={locabbr}"#&yearstart={yrstrt}&yearend={yrend}
query_url = f"{base_url}?{filters}&$limit={limit}&$$app_token={api_token}"

#Get data and convert to data frame
sample_question_data = requests.get(query_url).json()


df2=pd.DataFrame(sample_question_data)
print(len(df2))
df2
```
